```python
import torch
import torch.nn as nn
import torchvision
import torchvision.transforms as transforms
import torch.optim as optim

if torch.cuda.is_available():
    device = 'cuda'
else: 
    device = 'cpu'
    
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1=nn.Linear(28*28,100) 
        self.gc2=nn.Linear(100,10)
        
    def forward(self,x):
        input=torch.flatten(x,start_dim=1)
        x=torch.relu(self.fc1(input))
        x=torch.relu(self.gc2(x))
        return x
    
    
max_epoch=100
batch_size=32

transforms_processor=transforms.Compose([transforms.ToTensor()])
train_dat=torchvision.datasets.MNIST(root='./data', train=True, transform=transforms_processor, download=True)
train_dat_loader=torch.utils.data.DataLoader(train_dat, batch_size=batch_size, shuffle=True)

test_dat=torchvision.datasets.MNIST(root='./data', train=True, transform=transforms_processor, download=True)
test_dat_loader=torch.utils.data.DataLoader(train_dat, batch_size=batch_size, shuffle=True)



model=Net()
model.to(device)

loss_fn=nn.CrossEntropyLoss()
optimizer=optim.SGD(model.parameters(), lr=0.001, momentum=0.9)

def train(epoch):
    train_cnt=0
    for e in range(epoch):
        for i, (dat, labels) in enumerate(train_dat_loader):
            dat=dat.to(device)
            labels=labels.to(device)
            optimizer.zero_grad()
            outputs=model(dat)
            loss=loss_fn(outputs,labels)
            loss.backward()
            optimizer.step()
            
            train_predict=torch.max(outputs,dim=1)[1]
            train_cnt+=torch.eq(train_predict,labels.to(device)).sum().item()
        train_accuracy=train_cnt/len(train_dat)
            
        model.eval()
        test_accuracy_cnt=0
        best_test_acc=0.0
        with torch.no_grad():
            for i, (dat, labels) in enumerate(test_dat_loader):
                tedt_outputs=model(dat.to(device))
                pred=torch.max(tedt_outputs,dim=1)[1]
                test_accuracy_cnt+=torch.eq(labels.to(device),pred).sum().item()
                
            acc=test_accuracy_cnt/len(test_dat)
            if (acc>best_test_acc):
                torch.save(model.state_dict(),'./model.pth')
                best_test_acc=acc  
            print('epoch:%d train accuracy:%.3f test accuracy:%.3f' %(e,train_accuracy,acc))
            train_cnt=0
            train_accuracy=0.0
            best_test_acc=0.0
                
train(max_epoch)         
```
